### Importing the library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Loading the Data set

```python
df = pd.read_csv('Student_performance_data _.csv')
df
```

### Statical Analysis

```python
df.describe()
```

```python
# Correlation Matrix
correlation_matrix = df.corr()
print(correlation_matrix)
```

```python
# Seting the style of plots
sns.set(style='whitegrid')

#plotting distribution of key variable
plt.figure(figsize=(20,12))
```

```python
#Distribution of age
plt.subplot(2,2,1)
sns.histplot(df['Age'])
plt.title('Age Distribution')
plt.show()
```

```python
#Distribution of Study time weekly
plt.subplot(2,3,2)
sns.histplot(df['StudyTimeWeekly'], kde=True)
plt.title('Study Time Weekly Distribution')
plt.show()
```

```python
# Distribution of GPA
plt.subplot(2, 3, 3)
sns.histplot(df['GPA'], kde= True)
plt.title("GPA Distribution")
plt.show()
```

```python
# Distribution of Abesence 
plt.subplot(2,3,4)
sns.histplot(df['Absences'], kde=True)
plt.title('Abesence Distribution')
plt.show()
```

```python
# Distribution of Parental Support
plt.subplot(2,3,5)
sns.histplot(df['ParentalSupport'], kde=True)
plt.title('Parental Support Distribution')
plt.show()
```

```python
# Distribution of Grade Class
plt.subplot(2, 3, 6)
sns.histplot(df['GradeClass'], kde=True)
plt.title('Grade Class Distribution')
plt.show()
```

### Correlation Heatmap

```python
plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

### Pair Plot

#### It will show the relationshop between pairs of variable

```python
sns.pairplot(df[['GPA', 'Absences', 'StudyTimeWeekly', 'ParentalSupport', 'Gender']], hue='Gender', palette='coolwarm')
plt.suptitle('Pair Plot of key variable', y=1.02)
plt.show()
```

### Facet Grid

#### Facet Grid to show GPA distribution across different levels of parental support and gender

```python
g= sns.FacetGrid(df, row='ParentalSupport', col='Gender', margin_titles=True, height=3, aspect=1.5)
g.map(sns.histplot, 'GPA')
g.add_legend()
plt.suptitle('GPA Distribution by Parental Support and Gender', y=1.02)
plt.show()
```

### 3D Scatter Plot

#### 3D Scatter plot with GPA, StudyTimeWeekly, and Absences

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
sc = ax.scatter(df['StudyTimeWeekly'], df['Absences'], df['GPA'], c=df['GPA'], cmap='viridis', marker='o')

ax.set_xlabel('Study Time Weekly')
ax.set_ylabel('Absences')
ax.set_zlabel('GPA')
plt.title('3D Scatter Plot of GPA, Study Time Weekly, and Absences')
plt.colorbar(sc, ax=ax, label='GPA')
plt.show()
```
